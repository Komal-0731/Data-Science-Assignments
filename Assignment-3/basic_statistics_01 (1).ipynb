```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
df=pd.read_csv("S:/study/Data science/Assignment/sales_data_with_discounts.csv")
df
```

```python
#Identify numerical columns in the dataset.
select_numerical_col=df.select_dtypes(include=['int64','float64']).columns
select_numerical_col
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
df.describe ()
```

```python
print("Mode :",df["Volume"].mode())
print("---------------------------")
print("Mode :",df["Avg Price"].mode())
print("---------------------------")
print("Mode :",df["Total Sales Value"].mode())
print("---------------------------")
print("Mode :",df["Discount Rate (%)"].mode())
print("---------------------------")
print("Mode :",df["Discount Amount"].mode())
print("---------------------------")
print("Mode :",df['Net Sales Value'].mode())
```

```python
#Histograms:
        #Plot histograms for each numerical column.
df[select_numerical_col].hist(figsize=(12,8))
plt.title("Histogram of Numerical Columns")
plt.show()
```

```python
# Box Plots:
        # Create boxplots for numerical variables to identify outliers and the interquartile range.
df[select_numerical_col].plot.box(figsize=(12,8))
plt.title("Boxplot of Numerical Columns")
plt.show()
```

```python
#Bar Chart Analysis for Categorical Column:
        	#Identify categorical columns in the dataset.
        	#Create bar charts to visualize the frequency or count of each category.
categorical_column=df.select_dtypes(include=["object"]).columns
categorical_column
```

```python
for col in categorical_column:
    df[col].value_counts().plot(kind='bar')
    plt.title(f"{col} Distribution")
    plt.ylabel("Count")
    plt.show()
```
