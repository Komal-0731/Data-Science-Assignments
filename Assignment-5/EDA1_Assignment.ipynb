<h1>3.	Data Visualization:</h1>

<h5>Histogram</h5>

```python
# Histogram for LB column
plt.figure(figsize=(10,6))
plt.hist( df['LB'],bins=20)
plt.title('Histogram of LB')
plt.xlabel("LB")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histograms for all numerical columns
df.hist(figsize=(15,8), bins=20)
plt.suptitle("Histograms of Numerical Variables")
plt.show()
```

<h5>Boxplot</h5>

```python
# Boxplot for LB column
plt.figure(figsize=(10,6))
sns.boxplot(x=df['LB'])
plt.title('Boxplot of LB')
plt.show()
```

```python
# Boxplot for all numerical columns
plt.figure(figsize=(15,8))
plt.title("Boxplot of Numerical Variable")
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

<h2>Boxplot Observation</h2>

<h4>The boxplot visualizes the distribution and spread of numerical variables in the dataset.</h4>
<h4>Several variables contain outliers, represented by points outside the whiskers.</h4>
<h4>Variables such as LB, ASTV, ALTV, and Width show high variability and extreme values.</h4>
<h4>The median values are represented by horizontal lines inside each box.</h4>
<h4>Some features have compact distributions, while others show wider interquartile ranges (IQR).</h4>
<h4>Since this is medical data, the outliers may represent abnormal fetal conditions and are therefore important for analysis.</h4>

<h5>Scatter Plots </h5>

```python
# Scatter plot between LB and AC

plt.figure(figsize=(8,6))

sns.scatterplot(x='LB', y='AC', data=df)

plt.title("Scatter Plot of LB vs AC")

plt.xlabel("LB")

plt.ylabel("AC")

plt.show()
```

```python
# Scatter plot between UC and FM

plt.figure(figsize=(8,6))

sns.scatterplot(x='UC', y='FM', data=df)

plt.title("Scatter Plot of UC vs FM")

plt.xlabel("UC")

plt.ylabel("FM")

plt.show()
```

```python
# Pairplot for multiple variables

sns.pairplot(df[['LB','AC','FM','UC','ASTV']])

plt.show()
```

<h5>correlation heatmaps </h5>

```python
# Correlation heatmap
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Violin plot

plt.figure(figsize=(10,6))

sns.violinplot(data=df[['LB','AC','FM','UC']])

plt.title("Violin Plot")

plt.show()
```
