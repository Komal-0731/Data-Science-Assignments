```python
import numpy as np
```

```python
#Find Theoretical Mean Cost

W = 1000+5*(600)
print("Theoretical Mean Cost = ",W)
```

<h5>State the Hypotheses statement</h5>

<h4>Null Hypothesis (H₀)</h4>
<h3>The weekly operating cost is not higher than the theoretical model.</h3>
         <h3>    μ = 4000</h3>
         
<h4>Alternative Hypothesis (H₁)</h4>
<h3>The weekly operating cost is higher than the theoretical model.</h3>
<h3>μ > 4000</h3>

```python
# Calculate Standard Deviation of Cost
sigma = 5 * 25
print(sigma)
```

```python
# Calculate Test Statistic (z)
x_bar = 3050
mu = 4000 
sigma = 125 
n=25

z = (x_bar-mu) / (sigma/np.sqrt(n))
print("Z Statistic =>",z)
```

```python
import scipy.stats as stats 
critical_value = stats.norm.ppf(0.95)
print("Critical z_value =>",round(critical_value,4))
```

```python
if z > critical_value:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

<h4>Conclusion</h4>

<h5>There is not enough evidence to support the restaurant owners' claim that the weekly operating costs are higher than the theoretical model suggests.</h5>
