<h1>ESTIMATION AND CONFIDENCE INTERVALS</h1>

```python
# data 

data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
n=15
```

```python
import numpy as np
import scipy.stats as stats
```

```python
mean=np.mean(data)
print('Sample Mean :',mean)
```

<h5>99% Confidence Interval Using Sample Standard Deviation</h5>

```python
std_sample = np.std(data,ddof=1)
print("Sample Standard Deviation :",std_sample)
```

```python
confidence = 0.99
t_value = stats.t.ppf((1 + confidence) / 2 , df=n-1)
print("t_value :",t_value)
```

```python
margin_error = t_value *(std_sample / np.sqrt(n))
print("Margin of Error : ",margin_error)
```

```python
lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval :")
print(lower , upper)
```

<h4>Interpretation</h4>
<h6>We are 99% confident that the true mean durability of the print-heads lies between 1.090 million and 1.387 million characters.</h6>

<h4>Why t-Distribution is Used?</h4>
<h6>The t-distribution is used because the sample size is small (n < 30) and the population standard deviation is unknown. It provides more accurate confidence intervals for small samples.</h6>

<h5>99% Confidence Interval Using Known Population Standard Deviation</h5>

```python
population_std = 0.2
```

```python
z_value = stats.norm.ppf(0.995)
print("z-value :",z_value)
```

```python
margin_error_z = z_value * (population_std / np.sqrt(n))
print("Margin of Error :",margin_error_z)
```

```python
lower_z = mean - margin_error_z
upper_z = mean + margin_error_z

print ("99% Confidence Interval using population standarddeviation :")
print(lower_z , upper_z)
```

<h4>Interpretation</h4>
<h6>We are 99% confident that the true mean durability lies between 1.106 million and 1.372 million characters when the population standard deviation is known.</h6>

<h3>Conclusion</h3>
<h5>The confidence interval estimates the range in which the true population mean is likely to exist. The t-distribution was used when the population standard deviation was unknown, while the z-distribution was used when the population standard deviation was known. Both intervals provide an estimate of the average durability of print-heads with 99% confidence.</h5>
