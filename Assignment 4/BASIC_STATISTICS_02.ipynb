```python
import numpy as np
import pandas as pd
```

```python
#1.	Load the patient dataset and show summary with info().
df=pd.read_csv("S:\study\Data science\Assignment\Patient_Data.csv")
df
```

```python
df.info()
```

```python
#Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
selected_columns_billing=df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
selected_columns_billing
```

```python
# Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
df_drop=df.drop(columns=['ReceptionistID','CheckInTime'])
df_drop
```

```python
#Use groupby to find total bill amount per department.
total_bill_per_department=selected_columns_billing.groupby('Department')['BillAmount'].sum().reset_index()
total_bill_per_department
```

```python
#Remove duplicate patient records based on PatientID.
Remove_dup_patient=df.drop_duplicates("PatientID")
Remove_dup_patient
```

```python
#Fill missing BillAmount values with the mean bill amount.

bill_amount=df['BillAmount'].mean()
df['BillAmount'] = df['BillAmount'].fillna(bill_amount)
df
```

```python
#Merge the billing dataset with patient dataset on PatientID.
marge_df=pd.merge(df,selected_columns_billing,on='PatientID')
marge_df
```

```python
#Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

new_billing_columns=pd.DataFrame({
    'InsuranceCovered': [4000, 2000, 3500],  
    'FinalAmount': [1000, 1000, 500] 
})
new_billing_columns
```

```python
#Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
concat_columns=pd.concat([marge_df,new_billing_columns],axis=1)
concat_columns
```
